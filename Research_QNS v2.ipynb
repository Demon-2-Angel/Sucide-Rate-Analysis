### Extra Question : Does High Income Affects Suiciderate ?


```python
df[["incomeperperson", "suicideper100th"]].corr()
```

```python
subset1 = df.loc[(df["incomeperperson"]>=2000) & (df["suicideper100th"]>8),["country", "incomeperperson", "suicideper100th"]]
subset1
```

```python
print("No. of Countries that have Income-per-person >= $2000 & suicideper100th > 8000 :\n")
print("Only",len(subset1), "such countries are there. ") 
```

```python
subset1.plot(x='incomeperperson', y='suicideper100th', kind='scatter')
plt.show()
```

```python
correlation1 = subset1['incomeperperson'].corr(subset1['suicideper100th'], method= 'pearson')
print(correlation1)
```

 This correlation represents that as income of a person increases, suciderate / 100 thousands decreases, whereas even in the graph you can observe that as income is increasing, sucide rate is decreasing.
